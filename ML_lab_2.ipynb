# **A2 : Data Imputation**

```python
# Mean : TT4, T4U
# Median : TSH, T3, FTI, TBG
# Mode : Sex
```

```python
# Filling values with median ( numeric and contain outliers)

median_TSH = df['TSH'].median()
df['TSH'].fillna(median_TSH, inplace=True)

median_T3 = df['T3'].median()
df['T3'].fillna(median_T3, inplace=True)

median_FTI = df['FTI'].median()
df['FTI'].fillna(median_FTI, inplace=True)

median_TBG = df['TBG'].median()
df['TBG'].fillna(median_TBG, inplace=True)
```

```python
# Filling values with mode (categorical attributes)

mode_sex = df['sex'].mode()[0]
df['sex'].fillna(mode_sex, inplace=True)
```

# **A3 : Data Normalization / Scaling**

```python
df.columns
```

```python
# Using Min-Max to Normalize

# Numeric columns to normalize
X = df[['Record ID','age','TSH','T3','TT4','T4U','FTI','TBG']]

# Minimum and Maximum range for scaling
Min_Range = 0
Max_Range = 1

# X.min() : Min value from each column
# X.max() : Max value from each column

# Min-Max scaling to the selected columns
Min_Max_Scaled = (X - X.min()) / (X.max() - X.min()) * (Max_Range - Min_Range) + Min_Range

# Replacing the original columns with the scaled values
df[X.columns] = Min_Max_Scaled

print("Scaled DataFrame:")
df
```

# **A4 : Similarity Measure**

```python
# Observation Vectors : 'Sex' and 'on thyroxine'

Observation_Vectors = df[['sex', 'on thyroxine']]
Observation_Vectors.head(20)
```

```python
# Converting Categorical Column into Numerical(Binary) Column using sklearn

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label_1 = le.fit_transform(Observation_Vectors['sex'])
Observation_Vectors = Observation_Vectors.drop('sex', axis='columns')
Observation_Vectors['sex'] = label_1


label_2 = le.fit_transform(Observation_Vectors['on thyroxine'])
Observation_Vectors = Observation_Vectors.drop('on thyroxine', axis='columns')
Observation_Vectors['on thyroxine'] = label_2

Observation_Vectors.head(20)
```

```python
# Jaccard Coefficient(JC = (f11) / (f01+ f10+ f11))

S1 = Observation_Vectors['sex']
S2 = Observation_Vectors['on thyroxine']

Intersection = sum((a and b)  for a, b in zip(S1, S2))
Union = sum((a or b) for a, b in zip(S1, S2))
JC = Intersection / Union

print("Jaccard Coefficient (JC):", JC)
```

```python
# Simple Matching Coefficient (SMC= (f11+ f00) / (f00 + f01+ f10+ f11))

Num_Matching = sum(a == b for a, b in zip(S1, S2))
Num_of_Attributes = len(S1)
SMC = Num_Matching / Num_of_Attributes

print("Simple Matching Coefficient (SMC):", SMC)
```

# **A5 : Cosine Similarity Measure**

```python
from numpy.linalg import norm

cosine = np.dot(S1,S2)/(norm(S1)*norm(S2))
print("Cosine Similarity:", cosine)
```

# **A6 : Heatmap Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

vectors = df.iloc[:20]

def J_C(vector1, vector2):
    intersection = len(set(vector1) & set(vector2))
    union = len(set(vector1) | set(vector2))
    return intersection / union

def S_M_C(vector1, vector2):
    Matching = len(set(vector1) & set(vector2))
    Total_Attributes = len(vector1)
    return Matching / Total_Attributes

jc_matrix = [[J_C(vector1, vector2) for vector2 in vectors] for vector1 in vectors]
smc_matrix = [[S_M_C(vector1, vector2) for vector2 in vectors] for vector1 in vectors]
```

```python
# Jaccard Coefficient Heatmap

plt.figure(figsize=(16, 9))
sns.heatmap(jc_matrix, annot=True, fmt=".2f", cmap="BuPu", xticklabels=False, yticklabels=False)
plt.title("Jaccard Coefficient Heatmap")
plt.show()
```

```python
# Simple Matching Coefficient Heatmap

plt.figure(figsize=(16, 9))
sns.heatmap(smc_matrix, annot=True, fmt=".2f", cmap="BuPu", xticklabels=False, yticklabels=False)
plt.title("Simple Matching Coefficient Heatmap")
plt.show()
```
