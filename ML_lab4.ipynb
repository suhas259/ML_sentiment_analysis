```python
def entropy(data):
    labels = [row[-1] for row in dataset]
    unique_labels, label_counts = np.unique(labels, return_counts=True)
    probabilities = label_counts / len(labels)
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy
entropy(dataset)
```
